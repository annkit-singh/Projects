```python
import numpy as np
import cv2
import os
```

```python
import faceRecognition as fr
print (fr)
```

```python
test_img=cv2.imread(r'C:\Users\BELIEVER\Downloads\Test_Image.jpg')  
```

```python
faces_detected,gray_img=fr.faceDetection(test_img)
print("face Detected: ",faces_detected)
```

```python
faces,faceID=fr.labels_for_training_data(r'C:\train') #Give path to the train-images folder which has both labeled folder as 0 and 1
face_recognizer=fr.train_classifier(faces,faceID)
face_recognizer.save(r'C:\Savefolder\trainingData.yml') #It will save the trained model. Just give path to where you want to save

name={0:"Aankit",1:"Rajat"} 
```

```python
for face in faces_detected:
    (x,y,w,h)=face
    roi_gray=gray_img[y:y+h,x:x+h]
    label,confidence=face_recognizer.predict(roi_gray)
    print ("Confidence :",confidence)
    print("label :",label)
    fr.draw_rect(test_img,face)
    predicted_name=name[label]
    fr.put_text(test_img,predicted_name,x,y)

resized_img=cv2.resize(test_img,(1000,700))

cv2.imshow("face detection ", resized_img)
cv2.waitKey(10)
cv2.destroyAllWindows
```
